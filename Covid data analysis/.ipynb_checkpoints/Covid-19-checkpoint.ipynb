# Pearson Correlation on COVID-19 Data

Ministry of Health and Family Welfare, Government of India, wanted to look at the relationship between the number of COVID-19 confirmd cases and the number of deaths associated with it. Use the attached csv file provided by WHO and do the following;

#### Research Question 1:

Is there a relationship between the number of confirmed cases and the number of deaths related to COVID-19?Hypothesis:   The 'Null Hypothesis' might be:

    H0: There is no correlation between the number of confirmed cases and the number of deaths (equivalent to saying r = 0)  

And an 'Alternative Hypothesis' might be:

    H1: There is a correlation between the number of confirmed cases and the number of deaths (equivalent to saying r ≠ 0),

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Covid-19_Data_WHO.csv")
```

H0: There is no correlation between the number of confirmed cases and the number of deaths (equivalent to saying r = 0)  
H1: There is a correlation between the number of confirmed cases and the number of deaths (equivalent to saying r ≠ 0),

```python
r = df['Confirmed Cases Per Day'].corr(df['Deaths Per Day'])
print(r)
```

r ≠ 0

H1 is true, there is a correlation between the number of confirmed cases and the number of deaths

```python
import matplotlib.pyplot as plt

plt.scatter(df['Confirmed Cases Per Day'], df['Deaths Per Day'])
plt.show()
```

A strong positive corelation can be identified from the graph

```python
import math
from scipy import stats
from scipy.special import factorial
from matplotlib import pyplot as plt

def likelihood(theta, n, x):
    return (factorial(n) / (factorial(x) * factorial(n - x))) * (theta ** x) * ((1 - theta)  (n - x))

def pprob(prior, posterior, n_occured, n_events):
    return pd.Series(map(lambda theta: likelihood(theta, n_events, n_occured), prior))

def generative_model(n_events, p):
    return np.random.binomial(n_events, p)

def ABC(n_occured, n_events, n_draws=1000):
    prior = pd.Series(sorted(np.random.uniform(0, 1, size=n_draws)))
    sim_data = [generative_model(n_events ,p) for p in prior]
    posterior = prior[list(map(lambda x: x == n_occured, sim_data))]
    posterior_probability = pprob(prior, posterior, n_occured, n_events)

    # let's see what we got
    f, ax = plt.subplots(1)
    ax.plot(prior, posterior_probability)
    ax.set_xlabel("Theta")
    ax.set_ylabel("Likelihood")
    ax.grid()
    ax.set_title("Likelihood of Theta for New Campaign")
    plt.show()

ABC(10, 16)
```
